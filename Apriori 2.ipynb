```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
df=pd.read_csv('store_data.csv',header=None)
df.head()
```

```python
df.fillna(0,inplace=True)
df.head()
```

```python
df.shape
```

```python
transactions=[]
for i in range(0,len(df)):
    transactions.append([str(df.values[i,j]) for j in range(0,20) if str(df.values[i,j])!='0'])
```

```python
transactions[0]
```

```python
transactions[1]
```

```python
rules=apriori(transactions,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2)
```

```python
Results=list(rules)
df_results=pd.DataFrame(Results)
df_results.head()
```

```python
support=df_results.support
```

```python
first_values=[]
second_values=[]
third_values=[]
fourth_values=[]
for i in range(df_results.shape[0]):
    single_list=df_results['ordered_statistics'][i][0]
    first_values.append(list(single_list[0]))
    second_values.append(list(single_list[1]))
    third_values.append(single_list[2])
    fourth_values.append(single_list[3])
    
```

```python
df_results['ordered_statistics'][0][0]
```

```python
lhs=pd.DataFrame(first_values)
rhs=pd.DataFrame(second_values)
confidence=pd.DataFrame(third_values,columns=['Confidence'])
lift=pd.DataFrame(fourth_values,columns=['Lift'])
```

```python
final_df=pd.concat([lhs,rhs,support,confidence,lift],axis=1)
final_df
```

```python
final_df.fillna(value=' ',inplace=True)
final_df.head()
```

```python
final_df.columns=['lhs',1,2,'rhs',3,'support','confidence','lift']
final_df.head()
```

```python
final_df['lhs']=final_df['lhs']+str(', ')+final_df[1]+str(', ')+final_df[2]
final_df['rhs']=final_df['rhs']+str(', ')+final_df[3]
final_df.head()
```

```python
final_df.drop(columns=[1,2,3],inplace=True)
final_df.head()
```
